```python
# Problem A1: equiprobability
# this takes about 5 minutes to run

import random, math

# Direct sampling for disks in a box
#
# disks are not allowed to overlap, overlapping configurations are rejected
# disks are not allowed to overlap, overlapping configurations are rejected
# once a configuration of N-disks is accepted, not overlapping, then it is returned
#
def direct_disks_box(N, sigma):
    condition = False
    while condition == False:
        
        # generate a random coordinate between [0,1-sigma] in x and y
        # this coordinate is where the disk of radius sigma is placed
        # *note: this coordinate always gives a disk that fits entirely in the unit box
        L = [(random.uniform(sigma, 1.0 - sigma), random.uniform(sigma, 1.0 - sigma))]

        # for N-1 more iterations
        for k in range(1, N):
            
            # generate another valid coordinate in unit box
            a = (random.uniform(sigma, 1.0 - sigma), random.uniform(sigma, 1.0 - sigma))
            
            # compute the minimume euclidean distance from this coordinate and all other coordinates
            # in the current vector L (which contains coordinates of valid disks)
            min_dist = min(math.sqrt((a[0] - b[0]) ** 2 + (a[1] - b[1]) ** 2) for b in L)
            
            # if the minimum distance is less than the diameter of a disk, then they are overlapping
            # so we break, rejecting adding this point to the current configuration and 
            # start over **completely**
            if min_dist < 2.0 * sigma:
                condition = False
                break
                # *note: if the code gets here we do not have an accepted configuration and enforce tabular rasa
                
            # otherwise, we accept the current point a and append to L
            else:
                L.append(a)
                condition = True
    return L



def direct_disks_box_multirun_equiprobability(n_runs,sigma=0.15,del_xy=0.05):
    #sigma = 0.15 # radius of a disk
    #del_xy = 0.05
    conf_a = ((0.30, 0.30), (0.30, 0.70), (0.70, 0.30), (0.70,0.70))
    conf_b = ((0.20, 0.20), (0.20, 0.80), (0.75, 0.25), (0.75,0.75))
    conf_c = ((0.30, 0.20), (0.30, 0.80), (0.70, 0.20), (0.70,0.70))
    configurations = [conf_a, conf_b, conf_c]
    hits = {conf_a: 0, conf_b: 0, conf_c: 0}
    for run in range(n_runs):
        x_vec = direct_disks_box(4, sigma)
        for conf in configurations:
            #print("conf = "+conf)
            condition_hit = True
            for b in conf:
                # for all x_vec
                # compute absolute distance from x & y components of b to this x_vec
                # extract max of these distances, which is the maximum amount b is from this x_vec in one dimension
                # after repeated for all x_vec, get the minimum distance
                # if this minimum distance is less than del_xy, then condition_b is true
                # ---> this implies we are within the red box, or location threshold of our 4 configurations
                # condition_b is then multiplied by the current condition_hit
                condition_b = min(max(abs(a[0] - b[0]), abs(a[1] - b[1])) for a in x_vec) < del_xy
                condition_hit *= condition_b
            if condition_hit:
                # update the hit count for this configuration, which will accumulate over all n_runs
                hits[conf] += 1

    # display the number of hits (over all n_runs) for each configuration
    for conf in configurations:
        print conf, hits[conf]
    return hits
```

```python
# Problem A2: equiprobability and markov chain algorithm
# * a repeat of A1 with markov chain instead of direct sampling

# update disk configuration L using markov chain algorithm 
# L = current disk configuration
# delta = step for markov chain algorithm
# sigma = disk radius
def markov_disks_box_update(L,delta,sigma):
        # choose one of the disks  coordinates randomly
        a = random.choice(L)
        
        # take a step +/- delta from position of said disk
        b = [a[0] + random.uniform(-delta, delta), a[1] + random.uniform(-delta, delta)]
        
        # for other c disks in L ( where c != a) compute their distance from eachother 
        # and find the minimum distance
        min_dist = min((b[0] - c[0]) ** 2 + (b[1] - c[1]) ** 2 for c in L if c != a)
        
        # is minimum of disk position after step less than sigma
        # or
        # is maximum greater than 1.0 - sigma?
        # if so, then the disk is too close to the box edges which is no bueno
        box_cond = min(b[0], b[1]) < sigma or max(b[0], b[1]) > 1.0 - sigma
        
        # if new disk position b is far enough from the edges,
        # and far enough away from other disks, 
        # then put it in a (which is really part of L)
        if not (box_cond or min_dist < 4.0 * sigma ** 2):
            a[:] = b
        return L

def markov_disks_box_multirun_equiprobability(n_runs,delta=0.1,del_xy=0.05):
    conf_a = ((0.30, 0.30), (0.30, 0.70), (0.70, 0.30), (0.70,0.70))
    conf_b = ((0.20, 0.20), (0.20, 0.80), (0.75, 0.25), (0.75,0.75))
    conf_c = ((0.30, 0.20), (0.30, 0.80), (0.70, 0.20), (0.70,0.70))
    configurations = [conf_a, conf_b, conf_c]
    hits = {conf_a: 0, conf_b: 0, conf_c: 0}
    sigma = 0.15 # disk radius
    
    # initial location of disks
    L = [[0.25, 0.25], [0.75, 0.25], [0.25, 0.75], [0.75, 0.75]]
    
    for run in range(n_runs):
        
        # update step using markov chain algorithm
        L=markov_disks_box_update(L,delta,sigma)
        
        # Evaluate if configuration L is to each configuration a-c
        for conf in configurations:
            condition_hit = True
            for b in conf:
                condition_b = min(max(abs(a[0] - b[0]), abs(a[1] - b[1])) for a in L) < del_xy
                condition_hit *= condition_b
            if condition_hit:
                hits[conf] += 1

    for conf in configurations:
        print conf, hits[conf]
    return hits
```

```python
# Problem A3: event driven disk simulation at integer time intervals

import math, pylab

# doesn't change from event_disks_box.py
def wall_time(pos_a, vel_a, sigma):
    if vel_a > 0.0:
        del_t = (1.0 - sigma - pos_a) / vel_a
    elif vel_a < 0.0:
        del_t = (pos_a - sigma) / abs(vel_a)
    else:
        del_t = float('inf')
    return del_t

# doesn't change from event_disks_box.py
def pair_time(pos_a, vel_a, pos_b, vel_b, sigma):
    del_x = [pos_b[0] - pos_a[0], pos_b[1] - pos_a[1]]
    del_x_sq = del_x[0] ** 2 + del_x[1] ** 2
    del_v = [vel_b[0] - vel_a[0], vel_b[1] - vel_a[1]]
    del_v_sq = del_v[0] ** 2 + del_v[1] ** 2
    scal = del_v[0] * del_x[0] + del_v[1] * del_x[1]
    Upsilon = scal ** 2 - del_v_sq * ( del_x_sq - 4.0 * sigma **2)
    if Upsilon > 0.0 and scal < 0.0:
        del_t = - (scal + math.sqrt(Upsilon)) / del_v_sq
    else:
        del_t = float('inf')
    return del_t

conf_a = ((0.30, 0.30), (0.30, 0.70), (0.70, 0.30), (0.70,0.70))
conf_b = ((0.20, 0.20), (0.20, 0.80), (0.75, 0.25), (0.75,0.75))
conf_c = ((0.30, 0.20), (0.30, 0.80), (0.70, 0.20), (0.70,0.70))
configurations = [conf_a, conf_b, conf_c]
hits = {conf_a: 0, conf_b: 0, conf_c: 0}
del_xy = 0.10
pos = [[0.25, 0.25], [0.75, 0.25], [0.25, 0.75], [0.75, 0.75]]
vel = [[0.21, 0.12], [0.71, 0.18], [-0.23, -0.79], [0.78, 0.1177]]
singles = [(0, 0), (0, 1), (1, 0), (1, 1), (2, 0), (2, 1), (3, 0), (3, 1)]
pairs = [(0, 1), (0, 2), (0, 3), (1, 2), (1, 3), (2, 3)]
sigma = 0.10
t = 0.0
n_events = 5000000 # much larger number of events than event_disks_box.py
for event in range(n_events):
    if event % 100000 == 0:
        print("number of events so far in "+str(t)+" seconds = "+str(event))

    # find time for next event
    wall_times = [wall_time(pos[k][l], vel[k][l], sigma) for k, l  in singles]
    pair_times = [pair_time(pos[k], vel[k], pos[l], vel[l], sigma) for k, l in pairs]
    next_event = min(wall_times + pair_times)
    t_previous = t

    # consider times at integer steps
    for inter_times in range(int(t + 1), int(t + next_event + 1)):

        # get time change based on current time step and previous time
        del_t = inter_times - t_previous

        # for each disk, update position based on time change
        for k, l in singles:
            pos[k][l] += vel[k][l] * del_t

        # previous time is not changed to this time interval (changes each iter)
        t_previous = inter_times

        # evaluate hits for each configuration
        #print(t_previous) # for finding interval of times configurations are analyzed
        for conf in configurations:
            condition_hit = True
            for b in conf:
                condition_b = min(max(abs(a[0] - b[0]), abs(a[1] - b[1])) for a in pos) < del_xy
                condition_hit *= condition_b
            if condition_hit:
                hits[conf] += 1

    # update time to be the time of next event
    t += next_event


    # find change in time from previous event
    del_t = t - t_previous

    # update position, and velocities base on the time change
    for k, l in singles:
        pos[k][l] += vel[k][l] * del_t
    if min(wall_times) < min(pair_times):
        collision_disk, direction = singles[wall_times.index(next_event)]
        vel[collision_disk][direction] *= -1.0
    else:
        a, b = pairs[pair_times.index(next_event)]
        del_x = [pos[b][0] - pos[a][0], pos[b][1] - pos[a][1]]
        abs_x = math.sqrt(del_x[0] ** 2 + del_x[1] ** 2)
        e_perp = [c / abs_x for c in del_x]
        del_v = [vel[b][0] - vel[a][0], vel[b][1] - vel[a][1]]
        scal = del_v[0] * e_perp[0] + del_v[1] * e_perp[1]
        for k in range(2):
            vel[a][k] += e_perp[k] * scal
            vel[b][k] -= e_perp[k] * scal

for conf in configurations:
    print conf, hits[conf]

print("total time that has passed = "+str(t)+" seconds")
```

```python
#Problem B1: equiprobability for observable=x-position
import random, pylab

def direct_disks_box(N, sigma):
    overlap = True
    while overlap == True:
        L = [(random.uniform(sigma, 1.0 - sigma), random.uniform(sigma, 1.0 - sigma))]
        for k in range(1, N):
            a = (random.uniform(sigma, 1.0 - sigma), random.uniform(sigma, 1.0 - sigma))
            min_dist_sq = min(((a[0] - b[0]) ** 2 + (a[1] - b[1]) ** 2) for b in L)
            if min_dist_sq < 4.0 * sigma ** 2:
                overlap = True
                break
            else:
                overlap = False
                L.append(a)
    return L


N = 4
sigma = 0.1197
n_runs = 1000000
histo_data = []
for run in range(n_runs):
    pos = direct_disks_box(N, sigma)
    for k in range(N):
        histo_data.append(pos[k][0])
pylab.hist(histo_data, bins=100, normed=True)
pylab.xlabel('x')
pylab.ylabel('frequency')
pylab.title('Direct sampling: x coordinate histogram (density eta=0.18)')
pylab.grid()
pylab.savefig('direct_disks_histo.png')
pylab.show()
```
